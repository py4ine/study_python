```python
# 시각화_산점도
# X축 = 독립변수(상대값)  y축 = 종속변수(내값)
# 점들이 직선에 가깝게 정렬 = 선형 관계  점들이 곡선 형태 = 비선형 관계

import matplotlib.pyplot as plt

# 예제 데이터
x = [1, 2, 3, 4, 5, 6]
y = [2.1, 2.9, 3.8, 5.2, 6.3, 7.5]

plt.scatter(x, y)
plt.title("linear Plot")
plt.xlabel("X value")
plt.ylabel("y value")
plt.show()
```

```python
# 상수 계수 (Correlation Coefficient) 확인
# 상관 계수는 변수 간의 선형 관계를 정략적으로 나타내는 값
# 상관 계수 값 +1=강한 양의 선형 관계, 0=선형 관계 없음, -1=강한 음의 선형 관계

import numpy as np

# 데이터 생성
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.1, 2.9, 3.8, 5.2, 6.3, 7.5])

# 상관 계수 계산
correlation = np.corrcoef(x, y)[0, 1]
print(f"상관 계수: {correlation}")
```

```python
# 잔차 분석
# 잔차가 렌덤하게 분포 = 선형 모델 적합
# 잔차가 특정 패턴(곡선 등) = 비선형 가능성

from sklearn.linear_model import LinearRegression

# 데이터 준비
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2.1, 2.9, 3.8, 5.2, 6.3, 7.5])

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(x, y)

# 예측값 생성
y_pred = model.predict(x)

# 잔차 계산
residuals = y - y_pred

# 잔차 시각화
plt.scatter(x, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("X value")
plt.ylabel("Residuals")
plt.show()
```

```python
# 결정 계수 (R2)값 확인
# R2 > 0.7 = 선형 관계가 유의미

from sklearn.metrics import r2_score

# 결정 계수 계산
r2 = r2_score(y, y_pred)
print(f"결정계수R2: {r2}")
```

```python
# 다항 피팅 (Polynomial Fitting)
# 다항 회귀를 사용해 데이터를 피팅하고 오차를 확인
# 선형 모델과 다항 모델의 성능을 비교하여 다항 모델 성능이 좋으면 비선형 가능성 높음

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 다항 회귀 준비
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# 모델 학습
model = LinearRegression()
model.fit(x_poly, y)

# 예측값 생성
y_poly_pred = model.predict(x_poly)

# 성능 비교
r2_poly = r2_score(y, y_poly_pred)
print(f"다항회귀R2: {r2_poly}")
```
